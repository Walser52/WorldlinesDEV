```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
## Sampling the Gaussian distribution using the Box-Muller Transform
```

```python
u = np.random.rand(1000)
u1 = u[:500]
u2 = u[500:]
```

```python
np.pi
```

```python
z0 = np.sqrt(-2*np.log(u1)) * np.cos(2 * np.pi* u1)
z1 = np.sqrt(-2*np.log(u1)) * np.sin(2 * np.pi* u2)
z = np.concatenate((z0,z1))
```

```python
plt.hist(z, bins = 10)
```
